```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart_cleaned.csv')
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='HeartDisease', data=df)
plt.title('Distribution of Heart Disease')
plt.savefig("output1.png")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.savefig("output2.png")
plt.show()
```

```python
df.hist(figsize=(15, 10), bins=30)
plt.suptitle("Histograms of All Features", fontsize=16)
plt.tight_layout()
plt.savefig("output3.png")
plt.show()
```

```python
numeric_cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR']
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numeric_cols])
plt.title("Boxplots of Numerical Features")
plt.savefig("output4.png")
plt.show()
```

```python
for col in numeric_cols:
    plt.figure(figsize=(6, 3))
    sns.kdeplot(data=df, x=col, fill=True)
    plt.title(f'Distribution of {col}')
    plt.savefig(f'kde_{col}.png')
    plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='HeartDisease', y='Age', data=df)
plt.title('Age Distribution by Heart Disease')
plt.savefig("output5.png")
plt.show()
```

```python
df = pd.read_csv("HeartPrediction.csv")

categorical_features = ['Sex', 'ExerciseAngina', 'ChestPainType', 'RestingECG', 'ST_Slope']

for i, col in enumerate(categorical_features, start=6):
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, hue='HeartDisease', data=df)
    plt.title(f'Heart Disease by {col}')
    plt.xticks(rotation=45)
    plt.savefig(f'output{i}.png')
    plt.show()
```

```python
selected = ['Age', 'MaxHR', 'Cholesterol', 'HeartDisease']  

sns.pairplot(df[selected], hue='HeartDisease', diag_kind='kde')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.savefig("output_pairplot.png")
plt.show()
```
